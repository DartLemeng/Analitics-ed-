## Функции и библиотеки

```python
import pandas as pd

import matplotlib.pyplot as plt

import numpy as np

import seaborn as sns
```

```python
#функция room_similar
#назначение - возвращение количества комнат на основании медианной площади в передаваемом столбце 
#принимаемые данные:
# - площад объекта для которого вычисляется количество комнат (total_area), тип данных - int, float
# - медианные площади для объектов с различным количеством комнат
#возвращаемые данные:
# - количество комнат объекта (отсутствует подсчет для 13, 17,18 к кв)

def room_similar (total_area, type_area):
    try: #конструкция позволяющая избежать ошибок (например с типами данных или с отрицательными значениями) 
        if 0 < total_area <= (type_area[1] + (type_area[2] - type_area[1])/2): #диапазон соответствующий 1 к кв
            return 1
        elif (type_area[1] + (type_area[2] - type_area[1])/2) < total_area <= (type_area[2] + (type_area[3] - type_area[2])/2): #диапазон соответствующий 2 к кв
            return 2
        elif (type_area[2] + (type_area[3] - type_area[2])/2) < total_area <= (type_area[3] + (type_area[4] - type_area[3])/2): #диапазон соответствующий 3 к кв
            return 3
        elif (type_area[3] + (type_area[4] - type_area[3])/2) < total_area <= (type_area[4] + (type_area[5] - type_area[4])/2): #диапазон соответствующий 4 к кв
            return 4
        elif (type_area[4] + (type_area[5] - type_area[4])/2) < total_area <= (type_area[5] + (type_area[6] - type_area[5])/2): #диапазон соответствующий 5 к кв
            return 5
        elif (type_area[5] + (type_area[6] - type_area[5])/2) < total_area <= (type_area[6] + (type_area[7] - type_area[6])/2): #диапазон соответствующий 6 к кв
            return 6
        elif (type_area[6] + (type_area[7] - type_area[6])/2) < total_area <= (type_area[7] + (type_area[8] - type_area[7])/2): #диапазон соответствующий 7 к кв
            return 7
        elif (type_area[7] + (type_area[8] - type_area[7])/2) < total_area <= (type_area[8] + (type_area[9] - type_area[8])/2): #диапазон соответствующий 8 к кв
            return 8
        elif (type_area[8] + (type_area[9] - type_area[8])/2) < total_area <= (type_area[9] + (type_area[10] - type_area[9])/2): #диапазон соответствующий 9 к кв
            return 9
        elif (type_area[9] + (type_area[10] - type_area[9])/2) < total_area <= (type_area[10] + (type_area[11] - type_area[10])/2): #диапазон соответствующий 10 к кв
            return 10
        elif (type_area[10] + (type_area[11] - type_area[10])/2) < total_area <= (type_area[11] + (type_area[12] - type_area[11])/2): #диапазон соответствующий 11 к кв
            return 11
        elif (type_area[11] + (type_area[12] - type_area[11])/2) < total_area <= (type_area[12] + (type_area[14] - type_area[12])/2): #диапазон соответствующий 12 к кв
            return 12
        elif (type_area[12] + (type_area[14] - type_area[12])/2) < total_area <= (type_area[14] + (type_area[15] - type_area[14])/2): #диапазон соответствующий 14 к кв
            return 14
        elif (type_area[14] + (type_area[15] - type_area[14])/2) < total_area <= (type_area[15] + (type_area[16] - type_area[15])/2): #диапазон соответствующий 15 к кв
            return 15
        elif (type_area[15] + (type_area[19] - type_area[15])/2) < total_area <= (type_area[16] + (type_area[19] - type_area[16])/2): #диапазон соответствующий 16 к кв
            return 16
        elif total_area >(type_area[16] + (type_area[19] - type_area[16])/2): #диапазон соответствующий 19 к кв
            return 19
    except:
        return f'Ошибка входных данных, тип данных {type(total_area)} вместо int, float или {type(type_area)} вместо Siries' #сообщение о неверных передаваемых функции типах данных
```
